## 1. Estadística Descriptiva
### Objetivo:
Introducir los conceptos básicos de estadística descriptiva y su utilidad en el análisis de datos. Aprender a calcular y comprender las medidas de tendencia central, dispersión y posición, utilizando funciones en Python.

### Puntos clave a tratar:
La **estadística descriptiva** resume y describe las características más importantes de un conjunto de datos. Las medidas fundamentales incluyen:

**Medidas de tendencia central**:
- **Media (Promedio)**: La media es la suma de todos los valores dividida por el número total de observaciones. 

  Fórmula:
$$
\mu = \frac{1}{n} \sum_{i=1}^{n} x_i
$$
Donde:
- \( $\mu$ \) es la media,
- \( $n$ \) es el número total de observaciones,
- \( $x_i$ \) son los valores de las observaciones.

**Mediana**: La mediana es el valor central cuando los datos están ordenados. Si el número de observaciones es impar, es el valor del medio; si es par, es el promedio de los dos valores centrales.

**Moda**: La moda es el valor que más se repite en el conjunto de datos.

**Medidas de dispersión**:
- **Desviación estándar**: Mide la dispersión de los datos respecto a la media. La fórmula es:

$$
\sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2}
$$
Donde:
- \( $\sigma$ \) es la desviación estándar,
- \( $\mu$ \) es la media,
- \( $x_i$ \) son los valores de las observaciones,
- \( $n$ \) es el número total de observaciones.

**Rango**: El rango es la diferencia entre el valor máximo y el mínimo de un conjunto de datos. Fórmula:
$$
\text{Rango} = \text{Max}(x) - \text{Min}(x)
$$

**Varianza**: La varianza es el cuadrado de la desviación estándar. Fórmula:

$$
\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2
$$


**Medidas de posición**:
- **Cuartiles**: 
Los cuartiles dividen el conjunto de datos ordenados en cuatro partes iguales:
  - **Q1 (primer cuartil)**: El valor que separa el 25% inferior de los datos.
  - **Q2 (mediana)**: El valor que separa el 50% de los datos.
  - **Q3 (tercer cuartil)**: El valor que separa el 75% inferior de los datos.
- **Percentiles**: Los percentiles dividen el conjunto de datos en 100 partes iguales.

#### Funciones en Python:
- **`mean()`**: Calcula la media.
- **`median()`**: Calcula la mediana.
- **`mode()`**: Calcula la moda.
- **`std()`**: Calcula la desviación estándar.
- **`quantile()`**: Calcula los cuartiles.
- **`describe()`**: Resumen estadístico de un conjunto de datos.

## 3.3.- Opciones recomendas.

| Opción                              | Ventajas                                                              | Desventajas                                                         | Cuándo usar                                                      |
|-------------------------------------|-----------------------------------------------------------------------|--------------------------------------------------------------------|------------------------------------------------------------------|
| **Eliminar datos nulos**            | Simplicidad, sin suposiciones.                                        | Pérdida de información, puede sesgar los resultados si los datos nulos no son aleatorios. | Si los datos nulos son pocos y distribuidos aleatoriamente.      |
| **Rellenar con la media**           | Conserva el tamaño del conjunto de datos, evita sesgos por eliminación. | Introduce suposiciones, distorsiona la variabilidad.               | Si los datos nulos son significativos y la variable tiene distribución simétrica. |
| **Rellenar con la mediana**         | Robusto a los outliers, mejor que la media en datos sesgados.         | Puede no ser adecuado para todas las situaciones.                   | Para datos sesgados o con valores atípicos.                      |
| **Rellenar con la moda**            | Bueno para variables categóricas.                                     | Puede no reflejar la variabilidad de los datos.                    | En variables categóricas donde la moda tiene sentido.            |
| **Interpolación o modelos predictivos** | Relleno más sofisticado, adecuado para series temporales o relaciones complejas. | Requiere más trabajo y puede no ser necesario si el dataset es simple. | En datos temporales o cuando se tiene una relación compleja entre las variables. |


### 3.3.1.- Eliminar filas con función dropna

```python
# Eliminar filas con valores nulos
df.dropna(inplace=True)  
```

```python
# Eliminar columnas con valores nulos
df.dropna(subset=['Edad'], inplace=True)  # Elimina filas donde 'Edad' es nulo
```

### 3.3.2.- Relleno de datos nulos

```python
# Rellena con la media de la columna
df['Edad'].fillna(df['Edad'].mean(), inplace=True)  # Rellenar con la media
```

```python
# Rellena con la mediana
df['Edad'].fillna(df['Edad'].median(), inplace=True)  # Rellenar con la mediana
```

```python
# Rellena con la moda
df['Edad'].fillna(df['Edad'].mode()[0], inplace=True)  # Rellenar con la moda
```

### 3.3.3.- Interpolación de datos de manera lineal o polinómica

```python
df['Edad'] = df['Edad'].interpolate()  # Rellenar con interpolación lineal
```

```python
df['Edad'] = df['Edad'].interpolate(method='polynomial', order=2)  # Interpolación polinómica de segundo orden
```

## 4. Visualización de Datos con Matplotlib
La visualización es clave para comprender mejor los datos.
Vamos a crear gráficos simples para explorar estos datos.

```python
df_codificado.info()
```

```python
# 4.1 Gráfico de barras

df_codificado['Sexo'].value_counts().plot(kind='bar', title='Distribución por Sexo',
                                           color=['lightblue', 'lightgreen'])
plt.ylabel('Número de pasajeros')
plt.xlabel('Sexo')
plt.title('Distribución de Pasajeros por Sexo')
plt.show()
```

```python
# Filtrar hombres
mujeres = df_codificado[df['Sexo'] == 'female']

# Histograma
mujeres['Edad'].dropna().plot(kind='hist', bins=20, alpha=0.5, density=True, label='Histograma (Hombres)')

# Curva de densidad KDE
sns.kdeplot(mujeres['Edad'].dropna(), color='green', lw=2, label='KDE (Hombres)')

# Etiquetas y título
plt.xlabel('Edad')
plt.ylabel('Densidad')
plt.title('Distribución de Edad - Solo Mujeres')
plt.legend()

# Mostrar gráfico
plt.show()
```

```python
# Filtrar hombres
hombres = df_codificado[df_codificado['Sexo'] == 'male']  # O 'hombre' si ya está traducido

# Histograma
hombres['Edad'].dropna().plot(kind='hist', bins=20, alpha=0.5, density=True, label='Histograma (Hombres)')

# Curva de densidad KDE
sns.kdeplot(hombres['Edad'].dropna(), color='blue', lw=2, label='KDE (Hombres)')

# Etiquetas y título
plt.xlabel('Edad')
plt.ylabel('Densidad')
plt.title('Distribución de Edad - Solo Hombres')
plt.legend()

# Mostrar gráfico
plt.show()
```

```python
# 4.2 Histograma 
# Graficar el histograma
df_codificado['Edad'].dropna().plot(kind='hist', bins=20, alpha=0.5, title='Distribución de la Edad de los Pasajeros', density=True)

# Graficar la curva de densidad (KDE)
sns.kdeplot(df_codificado['Edad'].dropna(), color='red', lw=2)

# Etiquetas y título
plt.xlabel('Edad')
plt.ylabel('Densidad')
plt.title('Distribución de la Edad de los Pasajeros')

# Mostrar el gráfico
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df_codificado['Edad'].dropna().plot(kind='box', 
                                patch_artist=True,  # Rellenar la caja con color
                                boxprops=dict(facecolor='lightblue', color='blue'),  # Color de la caja
                                whiskerprops=dict(color='green'),  # Color de los bigotes
                                flierprops=dict(marker='o', color='red', markersize=5),  # Color y estilo de los puntos atípicos
                                medianprops=dict(color='red', linewidth=2))  # Color y grosor de la mediana

# Etiquetas y título
plt.ylabel('Edad')
plt.title('Boxplot de la Edad de los Pasajeros')

# Añadir grid para mejor visibilidad
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
# 4.2 Gráfico de dispersión (Ejemplo: Edad vs Tarifa)
plt.scatter(df_codificado['Edad'], df['Tarifa'])
plt.title('Edad vs Tafifa')
plt.xlabel('Edad')
plt.ylabel('Tarifa')
plt.show()
```

```python
numeric_cols = df_codificado.select_dtypes(include=['float64', 'int64'])

# Calcular la matriz de correlación
correlation_matrix = numeric_cols.corr()

# Crear un heatmap para visualizar la correlación
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Título y etiquetas
plt.title('Matriz de Correlación - Titanic Dataset')
plt.show()
```
